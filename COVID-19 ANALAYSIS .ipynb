```python

from bokeh.plotting import figure, curdoc
#from bokeh.driving import linear
from bokeh.models import ColumnDataSource,HoverTool
#import random
#import csv
#from bokeh.io import output_notebook,output_file,show,save
#from datetime import datetime as dt
#from bokeh.models import DatetimeTickFormatter
from bokeh.layouts import layout,column,row

```

Bokeh is a data visualization library in Python that provides high-performance interactive charts and plots. Bokeh output can be obtained in various mediums like notebook, html and server. It is possible to embed bokeh plots in Django and flask apps. Bokeh provides two visualization interfaces to users:

```python
source=ColumnDataSource(dfi)


f=figure(x_axis_type='datetime',title="COVID DEATHS IN INDIA",x_axis_label='Date(MONTH/YEAR)',y_axis_label='NUMBER OF DEATHS',plot_height=600,plot_width=600,tools="crosshair,pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select,")
f.title.align = 'center'
f.line(x='date',y='new_deaths',source=source,color='red')
f.circle(x='date',y='new_deaths',source=source,color='blue')


f1=figure(x_axis_type='datetime',title="COVID CASES IN INDIA",x_axis_label='Date(MONTH/YEAR)',y_axis_label='NUMBER OF CASES',plot_height=600,plot_width=600,tools="crosshair,pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select,")
f1.title.align= 'center'
f1.line(x='date',y='new_cases',source=source,color='yellow')
f1.circle(x='date',y='new_cases',source=source,color='orange')
f.add_tools(HoverTool(
    tooltips=[
        ('date',   '@date{%F}'            ),
        ('total deaths', '@total_deaths'),
        ('new deaths', '@new_deaths'),
    ],

    formatters={
        '@date'      : 'datetime', 
    },))
f1.add_tools(HoverTool(
    tooltips=[
        ('date',   '@date{%F}'            ),
        ('total cases', '@total_cases'),
        ('new cases', '@new_casess'),
    ],

    formatters={
        '@date'      : 'datetime', 
    },))
    
 
layout = row(f,f1)


show(layout)
```
